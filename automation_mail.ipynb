```python
df = df.dropna(subset=['Email', 'Company'])
```

```python
import re

# Function to validate email addresses
def is_valid_email(email):
    return re.match(r'^[\w\.-]+@[\w\.-]+\.\w+$', email) is not None
```

```python
# Function to establish SMTP connection
def connect_smtp():
    server = smtplib.SMTP('smtp.gmail.com', 587)
    server.starttls()
    server.login(sender_email, sender_password)
    return server

# Start SMTP connection
server = connect_smtp()
```

```python
# Loop through each row
for index, row in df.iterrows():
    try:
        recipient_email = row['Email']  # Exact column name
        company_name = row['Company']  # Exact column name

        # Validate email
        if not is_valid_email(recipient_email):
            print(f"Skipping invalid email: {recipient_email}")
            continue

        # Constructing the Email
        msg = MIMEMultipart()
        msg['From'] = sender_email
        msg['To'] = recipient_email
        msg['Subject'] = subject

        # Email body
        body = f"""\
        Dear {company_name} Team,

        As a skilled data scientist with an MSc in Data Science from Vellore Institute of Technology and a BSc in Statistics from Guwahati University, I bring a strong academic foundation and practical experience to the field.

        During my tenure as a Data Scientist Intern at GeeksforGeeks, I contributed to data-driven content optimization strategies that resulted in a 150% increase in website traffic. I also developed an NLP-based bot, significantly enhancing review automation processes. In my role as a Data Analyst Intern at Snehalaya, I optimized data maintenance and reporting systems, achieving a 20% improvement in data accuracy and efficiency.

        My technical skills include expertise in Python, R, SQL, TensorFlow, and PyTorch, along with hands-on experience in machine learning applications like leukemia cell classification and startup ecosystem analysis. Furthermore, my research contributions have been recognized, with a publication in the Journal of Advanced Research in Applied Sciences and Engineering Technology.

        I am impressed by {company_name}'s commitment to innovation and excellence in the data-driven domain. I am eager to contribute to your team's success and leverage my skills to drive impactful results.

        Please find my resume attached for your review. I look forward to the opportunity to discuss how my expertise aligns with the needs of {company_name}.

        Thank you for your consideration.

        Best regards,
        Arundhuti Chakraborty
        Contact: 7002604559
        """
        msg.attach(MIMEText(body, 'plain'))

        # Attaching the resume
        with open(resume_file, 'rb') as attachment:
            part = MIMEBase('application', 'octet-stream')
            part.set_payload(attachment.read())
            encoders.encode_base64(part)
            part.add_header('Content-Disposition', f"attachment; filename=Arundhuti-Chakraborty-Data Scientist.pdf")
            msg.attach(part)

        # Send the email
        server.send_message(msg)
        print(f"Email successfully sent to {recipient_email}")

    except Exception as e:
        print(f"Error sending email to {recipient_email}: {e}")
```
